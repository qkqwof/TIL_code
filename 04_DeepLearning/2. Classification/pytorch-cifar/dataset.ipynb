```python
import torch
import os
import pdb
from PIL import Image
import torchvision.transforms as transforms

def label2idx(label):
    if label == 'airplane':
        return 0
    elif label == 'automobile':
        return 1
    elif label == 'bird':
        return 2
    elif label == 'cat':
        return 3
    elif label == 'deer':
        return 4
    elif label == 'dog':
        return 5
    elif label == 'frog':
        return 6
    elif label == 'horse':
        return 7
    elif label == 'ship':
        return 8
    elif label == 'truck':
        return 9
    else:
        raise ValueError('Invalid label')


class Dataset(torch.utils.data.Dataset):
    def __init__(self, root, transforms=transforms.ToTensor()):
        self.root = root
        self.transforms = transforms
        self.file_list = []     
        for filename in os.listdir(root):
            self.file_list.append(os.path.join(root, filename))

    def __len__(self):
        return len(self.file_list)
'''
이 부분에서 진짜 애 많이 먹는다.
디버깅할 때 오래 걸리는 부분이다.
나중에 개발하다보면 알 수 있음..
나머지 기능들은 다 torch.utils,data.Dataset 에서 상속 받아서 사용했다.
'''
    def __getitem__(self, idx):
        file_dir = self.file_list[idx]
        img = Image.open(file_dir)
        img = self.transforms(img)
        
        label = file_dir.split('/')[-1].split('_')[1].split('.')[0]
        # print(label)
        label = label2idx(label)
        # print(label)
        return img, label
```
