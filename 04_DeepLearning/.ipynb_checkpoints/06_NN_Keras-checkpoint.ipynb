       학습을 어떻게 할지를 정의

```python
import numpy as np
import keras
```

```python
# 1) 데이터셋 생성
x = np.array([0,1,2,3,4])
y = 2 * x + 1
```

```python
print(x)
print(y)
```

```python
# 2) 모델 생성 Sequential() -> 여기다가 model..layer를 하나씩 추가하는 방법
# model.add() 해서 레이어를 하나씩 추가한다.
# Dense()는 FullyConnectedLayer를 의미한다.
# Dense(출력갯수, input_shape=(1,))
model = keras.models.Sequential()
model.add(keras.layers.Dense(1,input_shape=(1,)))
```

```python
# keras.layers.Dense?
```

```python
# 3) compile() 사용해서 어떻게 학습할지를 정의
model.compile(optimizer=keras.optimizers.SGD(0.01),loss="mse", metrics=["accuracy"]) # 0.01 -> learning rate 지정
```

```python
# 4) fit() 학습을 진행
model.fit(x[:4], y[:4],epochs=1000,verbose=2) # training 속성, training 라벨
```

```python
# 5) predict() 예측하기
print("Targets : ", y[3:])
print("Predicts : ",model.predict(x[3:]).flatten()) # flatten -> 출력값을 1차원으로 펼치겠다
```

```python
# 6) evaluate() 평가하기
# evaluate 함수 -> 출력물 loss, accuracy
eval_value=model.evaluate(x[3:], y[3:])
eval_value
```

```python
loss_function=model.evaluate(x[3:], y[3:])
print("loss:{}, accuracy:{}".format(*loss_function))
```
