```python
# 상반기 주유소 판매 데이터 불러오기
first = pd.read_csv("C:\encore_PJY\py_workspace\data/2020년 상반기 주유소 판매가격.csv",encoding="cp949")
first
```

```python
second = pd.read_csv("C:\encore_PJY\py_workspace\data/2020년 하반기 주유소 판매가격.csv",encoding="cp949")
second
```

```python
# 상반기와 하반기 데이터를 상하로 결합
df = pd.concat([first,second])
df
```

```python
df["구"] = df["지역"].str.split(" ").str[1]
df
```

### 주유소 데이터 reshape

```python
df1 = df.pivot_table(index=["기간"],values=["고급휘발유","휘발유","경유","실내등유"]).reset_index()
df1
```

### 전기자동차 데이터 불러오기

```python
data = pd.read_csv("C:\encore_PJY\py_workspace\data/전기자동차.csv")
data
```

### 데이터 병합

```python
result = pd.merge(df1,data, on = "기간")
result
```

- **전기차의 보유 현황을 보기 위해 외부에서 월별 총 전기차 보유 등록대수에 대한 정보를 가져와서 기존 데이터와 병합하였다.**

```python
plt.figure(figsize=(12,3))
plt.subplot(131)
plt.scatter(x = result["경유"],y = result["전기"])
plt.subplot(132)
plt.scatter(x = result["고급휘발유"],y = result["전기"])
plt.subplot(133)
plt.scatter(x = result["휘발유"],y = result["전기"])
plt.show()
```

- **시각화의 결과를 보게 되면 처음에 전기자동차 차량 등록수와 경유, 고급휘발유, 휘발유 각각의 관계를 알아보고자 산점도 그래프로 그려보았다. 하지만 데이터의 관계가 명확하지가 않다.**

```python
plt.figure(figsize=(20,3))
plt.subplot(131)
sns.relplot(x=result["경유"],y=result["전기"])
plt.subplot(132)
sns.relplot(x=result["고급휘발유"],y=result["전기"])
plt.subplot(133)
sns.relplot(x=result["휘발유"],y=result["전기"])
plt.show()
```

```python
plt.figure(figsize=(20,3))
plt.subplot(131)
sns.regplot(x=result["경유"],y=result["전기"],fit_reg=True)
plt.subplot(132)
sns.regplot(x=result["고급휘발유"],y=result["전기"],fit_reg=True)
plt.subplot(133)
sns.regplot(x=result["휘발유"],y=result["전기"],fit_reg=True)
plt.show()
```

- **각각의 관계가 양의 관계인지 음의 관계인지가 명확하지가 않아서 regplot으로 추세선까지 그려보았다. 추세선을 보게 되면 전반적으로 음의 관계가 나온다.**

### 분석결과
    예상했던 분석의 결과는 전기차 보유 증가에 따라서 유류세 가격이 낮아진다는 것이다. 그러나 아무래도 2020년도의 월별 자료여서 데이터의 개수가 12개 밖에 되지 않고 전체적인 그래프 형태가 음의 관계라고 하기에는 부족한 부분이 있다. 또한 추세선으로부터 데이터들이 멀리 떨어져 있어서 약한 관계이기도 하다. 따라서 결과적으로 전기차 보유의 증가와 유류세 가격은 서로 연관이 없다고 할 수 있다.
